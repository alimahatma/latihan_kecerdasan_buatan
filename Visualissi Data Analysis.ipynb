```python
import matplotlib.pyplot as plt
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
mataKuliah = ["PBO","Kalkulus","Data Base","Bahasa Inggris","Desain Web"]
nilai = [80, 83, 98, 50, 35]
ax.bar(mataKuliah, nilai)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
merek = ["SAMSUNG","APPLE","XIOMI","OPPO","VIVO"]
penjualan = [300, 110, 433, 332, 234]
ax.pie(penjualan,labels=merek,autopct='%1.2f%%')
plt.title("Daftar Merek HP Paling Laku Cak ALI Cell")
plt.show()
```
